# Concatenation Of Data Frames

```python
# creating ind_weather data frame
ind_weather=pd.DataFrame({"city":["Bangalore","Delhi","Mumbai"],
                         "temperature":[30,45,32],
                         "Humidity":[80,60,78]})
ind_weather
```

```python
# creating us_weather data frame
us_weather=pd.DataFrame({"city":["Cambridge","Ethica","Atlanta"],
                         "temperature":[30,15,20],
                         "Humidity":[80,95,90]})
us_weather
```

```python
# concatenate or join these two DF
df=pd.concat([ind_weather,us_weather])
df
```

```python
# for uniformity of index 
df=pd.concat([ind_weather,us_weather],ignore_index=True)
df
```

```python
# default is always columnwise concatenation
# if we want row wise then put axis=1
df=pd.concat([ind_weather,us_weather],axis=1)
df
```

# Merge Data Frames

```python
# creating temperature DF
temperature_df=pd.DataFrame({"city":["Mumbai","Delhi","Bangalore","Hydeabad"],
                            "Temperature":[32,45,30,40]})
temperature_df
```

```python
# creating Humidity DF
humidity_df=pd.DataFrame({"city":["Mumbai","Delhi","Bangalore"],
                            "Humidity":[78,60,80]})
humidity_df
```

```python
# merging data frames i.e, like SQL JOIN
df=pd.merge(temperature_df,humidity_df,on="city")
df
```

```python
# if we need all cities in that case do OUTER JOIN
df=pd.merge(temperature_df,humidity_df,on="city",how="outer")
df
```

# Numerical Indexing (.loc & .iloc)

```python
# data frame with custom index
df=pd.DataFrame([1,2,3,4,5,6,7,8,9,19],index=[49,48,47,46,45,1,2,3,4,5])
df
```

```python
df.loc[45]  # it uses custom index if given otherwise default index
```

```python
df.iloc[45] # it only uses default index i.e, starting from 0
```

```python
df.iloc[4]
```

```python
# slicing
df.loc[:2]  # upto index 2
```

```python
# slicing
df.iloc[:2]   # only give first two element at index 0 and 1  [start:end(exclude)]
```
