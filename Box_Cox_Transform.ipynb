```python
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
X=np.random.beta(1,3,10000)
```

```python
sns.distplot(X)
```

```python
import pylab
stats.probplot(X,dist=stats.norm,plot=pylab)
pylab.show()
```

```python
Y,lamb=stats.boxcox(X)
sns.distplot(Y)
print(lamb)
```

```python
# Clearly see that boxcox transform,transform the variable as close as to normal.

stats.probplot(Y,dist=stats.norm,plot=pylab)
pylab.show()
```

```python
# Inverse boxcox transform

import scipy
Z=scipy.special.inv_boxcox(Y,0.4)
sns.distplot(Z)
```
