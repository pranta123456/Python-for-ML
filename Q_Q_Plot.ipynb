# Q-Q Plot

```python
import numpy as np
import pylab
import scipy.stats as stats
```

```python
std_normal=np.random.normal(loc=0,scale=1,size=1000)
for i in range(101):
    print(i,np.percentile(std_normal,i))
```

```python
X_samples=np.random.normal(loc=20,scale=5,size=5000)
stats.probplot(X_samples,dist="norm",plot=pylab)
pylab.show()
```

```python
X_samples_1=np.random.uniform(low=-1,high=1,size=1000)
stats.probplot(X_samples_1,dist="norm",plot=pylab)
pylab.show()
```
